# Data preparation

```python
# clean sex column in person dataset
# change from blanks to 'U' for unknown

df_person["SEX"] = df_person["SEX"].replace({"M": "Male",
                                             "F": "Female"})
df_person["SEX"] = df_person["SEX"].fillna("Unknown")
df_person["TAKEN_HOSPITAL"] = df_person["TAKEN_HOSPITAL"].fillna("Not know")
# Code Description CF Centre-front CR Centre-rear D Driver or rider LF Left-front LR Left-rear 
# NA Not applicable NK Not known OR Other-rear PL Pillion passenger PS Motorcycle sidecar passenger RR Right-rear
df_person["SEATING_POSITION"] = df_person["SEATING_POSITION"].replace({"CF": "Centre-front",
                                                                       "CR": "Centre-rear",
                                                                      "D": "Driver or rider",
                                                                      "LF": "Left-front",
                                                                      "LR": "Left-rear",
                                                                      "NA": "Not applicable",
                                                                      "NK": "Not known",
                                                                      "OR": "Other-rear",
                                                                      "PL": "Pillion passenger",
                                                                      "PS": "Motorcycle sidecar passenger",
                                                                      "RR": "Right-rear"})
# Code Description 1 Seatbelt worn 2 Seatbelt not worn 3 Child restraint worn 4 Child restraint not worn 
# 5 Seatbelt/restraint not fitted 6 Crash helmet worn 7 Crash helmet not worn 8 Not appropriate 9 Not known
df_person["HELMET_BELT_WORN"] = df_person["HELMET_BELT_WORN"].replace({1: "Seatbelt worn",
                                                                       2: "Seatbelt not worn",
                                                                      3: "Child restraint worn",
                                                                      4: "Child restraint not worn",
                                                                      5: "Seatbelt/restraint not fitted",
                                                                      6: "Crash helmet worn",
                                                                      7: "Crash helmet not worn",
                                                                      8: "Not appropriate",
                                                                      9: "Not known"})

df_person["EJECTED_CODE"] = df_person["EJECTED_CODE"].replace("9", None)
# df_person["AGE_GROUP"] = df_person["AGE_GROUP"].replace("5/12/2025", "Unknown")
df_person["TAKEN_HOSPITAL"] = df_person["TAKEN_HOSPITAL"].replace("Y", "Yes").replace("N", "No")

# A Australian Capital Territory B Commonwealth D Northern Territory 
# N New South Wales O Overseas Q Queensland S South Australia T Tasmania 
# V Victoria W Western Australia Z Not known _ Not available (Blank value entered)

df_person["LICENCE_STATE"] = df_person["LICENCE_STATE"].replace({"A": "Australian Capital Territory",
                                                                 "B": "Commonwealth",
                                                                "D": "Northern Territory",
                                                                "N": "New South Wales",
                                                                "O": "Overseas",
                                                                "Q": "Queensland",
                                                                "S": "South Australia",
                                                                "T": "Tasmania",
                                                                "V": "Victoria",
                                                                "W": "Western Australia",
                                                                "Z": "Not known"})

# cleaned person dataset
cleaned_person = df_person
cleaned_person.head(20)
# cleaned_person.to_csv('cleaned_person.csv', index=False)
```
