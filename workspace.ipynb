```python
# clean sex column in person dataset
# change from blanks to 'U' for unknown

df_person["SEX"] = df_person["SEX"].replace(" ", "U")
df_person["AGE_GROUP"] = df_person["AGE_GROUP"].replace(" ", "Unknown")
df_person["TAKEN_HOSPITAL"] = df_person["TAKEN_HOSPITAL"].replace(" ", "Not know")
df_person["EJECTED_CODE"] = df_person["EJECTED_CODE"].replace("9", None)
# df_person

# cleaned person dataset
cleaned_person = df_person
cleaned_person.to_csv('cleaned_person.csv', index=False)
```
